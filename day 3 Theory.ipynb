#  Decision Trees

**Types of Decision Trees**

1. Classification DT: This will determine a set of logical if-then conditions to classify the problem
   eg: type flower

2. Regression DT:It is used when we have numeric/continous data
    eg: measure height  of the son and father

### Entropy: It is the measure of randomness in dataset(It measure purity in data split)

### Information Gain: It is the measure of decrease in entropy after the dataset is split.

High Entropy(E1)------> Low Entropy(E2)

Info_Gain=E1-E2

**```Entropy(E) = E(i=1 to k)(P(value)*log2(p(value)))```**

k=no of values in dataset
p(value)=Percentage of value

# Support  Vector Machine(svm):

To find a hyperplane in an N-dimensional space(N-No of features) such that it distinctly classifies the data points.

1. **Support Vectors:** These are the points where it passing the marginal planes

2. **Hyperplanes:** It is line which seperate the two different classes.

3. **Marginal Planes:** The points which passes through support vectors

4. **Marginal Distance:** It is the distance between two marginal Planes

5. **Linearly Seperable:** It is the 2D plane

6. **Non-Linearly Seperable:**
        SVM Kernel: The main aim this is to convert low dimensional(2D) to high dimensional(3D)
